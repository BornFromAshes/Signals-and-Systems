# Q1
It can be easily seen from the graph that the period is equal to 4 seconds. We can also get the period by dividing 2*pi by pi and pi/2 and find the lowest common multiple between them. LCM (4,2) = 4

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.linspace(-10, 10, 1000)
x = np.sin(np.pi/2 * t) + 2 * abs(np.cos(np.pi * t))

plt.plot(t, x)
plt.title(r'x(t) = sin(pi/2 t) + 2|cos(pi t)|')
plt.xlabel(r't (in s)')
plt.grid()
```

# Q2
Period is the same as the previous question, the question also specified that the steps for this graph is 0.01, but to enhance the visibility I changed it to 0.1. You can change the 0.01 in np.arange().

```python
n = np.arange(-10, 10.0001, 0.1)
x = np.sin(np.pi/2 * n) + 2 * abs(np.cos(np.pi * n))

markerline, stemline, baseline = plt.stem(n, x)
plt.title(r'x[n] = sin[pi/2 n] + 2|cos[pi n]|')
plt.setp(stemline, linewidth = 1)
plt.setp(markerline, markersize = 0.0001)
plt.xlabel('n')
```

# Q3
Period = LCM(pi/omega, 2\*pi/omega) = 2\*pi/omega  
As it is apparent from the graphs below and considering the calculated period, As the amount of omega increases, the period decreases.


```python
t = np.linspace(-10, 10, 1000)
n = np.arange(-10, 10.0001)
figure, axis = plt.subplots(9, 2)
for omega in range(9):
  cont_sig_x = np.sin(2 * omega/3 * t) + np.cos(omega/3 * t)
  axis[omega,0].plot(t, cont_sig_x)

  disc_sig_x = np.sin(2 * omega/3 * n) + np.cos(omega/3 * n)
  markerline, stemline, baseline = axis[omega,1].stem(n, disc_sig_x)
```
