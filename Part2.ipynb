```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_ak(k, w, T1):
    if k == 0:
        return 2*T1/3
    else:
        return np.sin(k * w * T1) / (k * np.pi)

def generate_fourier_series(r, w, T1, T, num_points=1000):
    time_points = np.linspace(-5, 5, num_points)
    signal = np.zeros_like(time_points)

    for k in range(-r, r+1):
        ak = calculate_ak(k, w, T1)
        signal += ak * np.cos(2 * np.pi * k * time_points / T)

    return time_points, signal

def main():
    r = int(input("Enter r: ")) # Choose an appropriate value for r
    w = 2 * np.pi / 3
    T1 = 1
    T = 3  # T is the period

    time_points, reconstructed_signal = generate_fourier_series(r, w, T1, T)

    # Display the result
    plt.plot(time_points, reconstructed_signal, label='Reconstructed Signal')
    plt.title('Fourier Series Signal Reconstruction')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```

Gibbs phenomenon is a phenomenon that is observed in the Fourier series representation of a function. When a function is approximated by its Fourier series, which is the sum of the sine and cosine functions, the presence of a jump discontinuity in the original function can lead to an overshoot or "ring" near the discontinuity in the Fourier series approximation.

In particular, if a function has a jump discontinuity, the Fourier series tends to overestimate the jump at the discontinuity. This overshoot occurs even as the number of terms in the Fourier series increases. As more terms are added, the overshoot does not decrease. Instead, it concentrates near the discontinuity.

The Gibbs phenomenon is characterized by continuous oscillations (bells) close to discontinuity in the Fourier series approximation, and these oscillations do not disappear with increasing the number of terms in the series. Overshoot is always about 9% of the jump in range.

And yes it clearly happened in this example.
